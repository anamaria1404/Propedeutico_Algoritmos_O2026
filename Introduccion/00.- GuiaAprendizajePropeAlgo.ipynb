## 📖 Descripción de la Asignatura

La ciencia de datos es una disciplina que exige comprender elementos de diversas áreas de conocimiento que incluyen las matemáticas, probabilidad y programación, además te tener conocimientos específicos del área de donde se desprende la información.

En este contexto, el estudiante que opte por iniciar una especialización en ciencia de datos debe tener bases sólidas para aprovechar al máximo las técnicas modernas del análisis de datos.

El propedéutico de programación e implementación de algoritmos revisa estos conocimientos base para lograr que el estudiante pueda insertarse de una mejor manera en un posgrado cuyo objetivo es dar herramientas de frontera en ciencia de datos.
<hr>

## 🎯 Objetivos del Curso
| Objetivo General |
|------------------|
Fortalecer las competencias fundamentales de programación e implementación de algoritmos en los estudiantes de nuevo ingreso, proporcionando las bases computacionales y matemáticas necesarias para cursar con éxito un posgrado en ciencia de datos.| 
 
| Objetivos Específicos |
----------------------|
|<br>- **Fundamentos de programación:** Implementar programas estructurados en Python que apliquen correctamente variables, estructuras de control, funciones y manejo de errores como base para el desarrollo de soluciones analíticas.</br> 
<br>- **Estructuras de datos:** Seleccionar y utilizar estructuras de datos apropiadas (listas, diccionarios, matrices, dataframes) según el tipo de problema, optimizando el acceso y manipulación de información.</br> 
<br>- **Manipulación y exploración de datos:** Aplicar técnicas de limpieza, transformación y análisis exploratorio sobre conjuntos de datos reales utilizando herramientas estándar del ecosistema de ciencia de datos (Pandas, Matplotlib).</br>|
