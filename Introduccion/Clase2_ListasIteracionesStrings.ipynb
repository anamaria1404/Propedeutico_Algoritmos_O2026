## 2.1 Ciclos `for`

Las iteraciones son una manera de ejecutar cierto bloque de código repetidamente:

````python
# Planetas, de nuevo
planetas
````

````python
# Imprimir todos los planetas en la misma línea
for i in planetas:
    print(i)
````

````python
for i in planetas:
    print(i,end=' ')
````

Para construir un ciclo `for`, se debe especificar:

- el nombre de la variable que va a iterar (planeta),

- el conjunto de valores sobre los que va a iterar la variable (planetas).

Se usa la palabra `in`, en este caso, para hacerle entender a Python que *planeta* va a iterar sobre *planetas*.

El objeto a la derecha de la palabra `in` puede ser cualquier objeto **iterable**. Básicamente, un iterable es cualquier arreglo (listas, tuplas, conjuntos, arreglos de numpy, series de pandas...).

````python
mult = (2,3,4,4,5,8,9,10)
mult
````

````python
# Multiplicación como ciclo
x = 1
for i in mult:
    x = i*x
print(f'El producto es: {x}')
````

````python
# Código aquí
def fibonacci(n):
    '''Función que regresa los primeros n números de la sucesión de Fibonacci'''
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
````

````python
def fibonacci(n):
    '''Función que regresa los primeros n números de la sucesión de Fibonacci'''
    a, b = 0, 1
    elementos = [a,b]
    for i in range(2,n):
        c = a + b
        elementos.append(c)
        a = b
        b = c
    return elementos
````

````python
fibonacci(7)
````

````python
def fibonacci2(n):
    '''Función que regresa los primeros n números de la sucesión de Fibonacci'''
    a, b = 0, 1
    elementos = [a,b]
    for i in range(n-2):
        c = a + b
        elementos.append(c)
        a = b
        b = c
    return elementos
````

````python
def fibonacci3(n):
    fi = [0,1,1]
    if n == 0:
        return []
    elif n == 1:
        return [0]
    elif n == 2:
        return [0,1]
    elif n>2:
        for _ in range(2, n-1):
            fi.append(fi[-1] + fi[-2])
        return fi
````

## 2.2 Ciclos `while`

Son otro tipo de ciclos en Python, los cuales iteran hasta que cierta condición deje de cumplirse.

Por ejemplo:

````python
i = 20
while i >= 10:
    print(i)
    i -= 1
#para que no se vuelva un ciclo infinito, es necesario modificar la variable de control dentro del ciclo, en este caso i, para que eventualmente deje de cumplir la condición del while.
````

El argumento de un ciclo `while` se evalúa como una condición lógica, y el ciclo se ejecuta hasta que dicha condición sea **False**.

## Pausa: Recursión

Una manera adicional de ejecutar iteraciones se conoce como *recursión*, y sucede cuando definimos una función en términos de sí misma.

Por ejemplo, el $n$-ésimo número de la secuencia de Fibonacci, recursivamente sería:

````python
def fib_recursivo(n): 
    if n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fib_recursivo(n-1) + fib_recursivo(n-2) 
        
````

````python
fib_recursivo(5)
````

````python
def factorial(n):
    if n == 0:
        return 1
    elif n == 1:
        return 1
    else:
        return factorial(n-1)*n
````

````python
factorial(5)
````

Ejemplo: escribir la siguiente función usando un ciclo for.

````python
def cuantos_neg(lista_iterable):
    cnt_neg = 0
    for num in lista_iterable:
        if num < 0:
            cnt_neg += 1
    return cnt_neg
````

````python
def cuantos_negativos(lista_iterable):
    return sum([1 for num in lista_iterable if num < 0])
````

````python
cuantos_neg([-1, 0, 2, -5, 3, -7, 8])
````

````python
def negativos(lista_iterable):
    return [num for num in lista_iterable if num < 0]
````

````python
negativos([-1, 0, 2, -5, 3, -7, 8])
````

### Navegando entre listas, tuplas, diccionarios: `zip`

````python
l1 = ['nombre' , 'edad', 'estado']
l2 = ['juan', 25, 'Jalisco' ]
````

¿Cómo puedo asociar estos valores en un diccionario? Con `zip`:

````python
# Primero, obtener la lista de pares
list(zip(l1,l2))
````

````python
# Después obtener diccionario de relaciones
dict(list(zip(l1,l2)))
````
