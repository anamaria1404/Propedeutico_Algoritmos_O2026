## 🐍 Un poco de historia sobre Python

- Python fue creado a finales de los 80 por **Guido van Rossum** en el Centro para las Matemáticas y la Informática (CWI) en los Países Bajos, como sucesor del lenguaje ABC.
- El nombre "Python" proviene de la afición de su creador por los humoristas británicos **Monty Python**.
- Es un software libre, de código abierto, regido por la [Python Software Foundation License](https://es.wikipedia.org/wiki/Python_Software_Foundation_License).
- Es un **lenguaje interpretado** que prioriza la legibilidad y la simplicidad del código.

<div style="background-color:#f9f9f9; border-left:4px solid #000047; padding:10px;">

## 📓 ¿Qué es un Jupyter Notebook?

Un Jupyter Notebook es un documento interactivo que combina código ejecutable, texto enriquecido (Markdown), ecuaciones matemáticas y visualizaciones. Es ideal para prototipar, analizar datos y compartir resultados.

**Tipos de celdas:**
- **Código:** Ejecuta instrucciones en Python y muestra la salida.
- **Markdown:** Permite escribir texto, ecuaciones y agregar imágenes.

**Ejemplo de celda de código:**

```python
# Ejemplo: suma de dos números
a = 5
b = 3
print(f"La suma de {a} + {b} es {a + b}")
```
