## ¿Cómo funciona el aprendizaje supervisado?

1. **Datos etiquetados:** Se dispone de un conjunto de datos donde cada ejemplo tiene una entrada (features) y una salida conocida (etiqueta o valor).
2. **Entrenamiento:** El algoritmo aprende una función que mapea las entradas a las salidas.
3. **Predicción:** Una vez entrenado, el modelo puede predecir la salida para nuevas entradas no vistas.

### Tipos de problemas
- **Clasificación:** La salida es una etiqueta discreta (por ejemplo, spam/no spam, tipo de flor).
- **Regresión:** La salida es un valor continuo (por ejemplo, precio de una casa, temperatura).

## Ejemplos de algoritmos de aprendizaje supervisado

| Algoritmo | Tipo | Descripción |
|-----------|------|-------------|
| Regresión Lineal | Regresión | Predice valores continuos ajustando una línea recta a los datos. |
| Regresión Logística | Clasificación | Predice probabilidades para clasificación binaria. |
| k-Nearest Neighbors (k-NN) | Clasificación/Regresión | Predice la etiqueta según los vecinos más cercanos. |
| Árboles de Decisión | Clasificación/Regresión | Utiliza reglas en forma de árbol para tomar decisiones. |
| Random Forest | Clasificación/Regresión | Conjunto de árboles de decisión para mejorar precisión. |
| Support Vector Machines (SVM) | Clasificación/Regresión | Encuentra el hiperplano óptimo que separa las clases. |
| Redes Neuronales | Ambos | Modelos inspirados en el cerebro humano, capaces de aprender patrones complejos. |

# Regresión Lineal

> El **ajuste de curvas** es el proceso de construir una curva (función), que sea el mejor ajuste a una serie de puntos. Las curvas ajustadas pueden ser usadas como asistencia en la visualización de datos, para inferir valores de una función donde no hay datos disponibles, y para resumir la relación entre variables.

**Referencia**:
- https://en.wikipedia.org/wiki/Curve_fitting

La regresión lineal simple es un método estadístico para modelar la relación entre una variable independiente $x$ y una variable dependiente $y$.

La ecuación de la recta es:

$$y = mx + b$$

o de forma general:

$$y = \beta_0 + \beta_1 x$$

donde:
- $y$: variable dependiente
- $x$: variable independiente
- $\beta_0$: intercepto
- $\beta_1$: pendiente

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Generar datos sintéticos
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Visualizar los resultados
plt.scatter(X, y, color='blue', label='Datos')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Datos')
plt.legend()
plt.show()
```

```python
# Ajustar el modelo de regresión lineal
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Visualizar los resultados
plt.scatter(X, y, color='blue', label='Datos')
plt.plot(X, lin_reg.predict(X), color='red', label='Recta ajustada')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()

print(f'Intercepto (β₀): {lin_reg.intercept_[0]:.2f}')
print(f'Pendiente (β₁): {lin_reg.coef_[0][0]:.2f}')
```

# Regresión lineal múltiple

La regresión lineal múltiple extiende la regresión simple para modelar la relación entre una variable dependiente $y$ y varias variables independientes $x_1, x_2, \dots, x_n$.

La ecuación general es:

$$y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \dots + \beta_n x_n$$

donde:
- $y$: variable dependiente
- $x_1, x_2, \dots, x_n$: variables independientes
- $\beta_0$: intercepto
- $\beta_1, \beta_2, \dots, \beta_n$: coeficientes

```python
# Generar datos sintéticos para regresión múltiple
np.random.seed(1)
X_multi = np.random.rand(100, 3)
# Coeficientes reales: 5, -2, 3
y_multi = 2 + 5*X_multi[:,0] - 2*X_multi[:,1] + 3*X_multi[:,2] + np.random.randn(100)
```

```python
# Ajustar el modelo de regresión múltiple
lin_reg_multi = LinearRegression()
lin_reg_multi.fit(X_multi, y_multi)
```

```python
# Visualizar los coeficientes
print(f'Intercepto (β₀): {lin_reg_multi.intercept_:.2f}')
for i, coef in enumerate(lin_reg_multi.coef_):
    print(f'Coeficiente β{i+1}: {coef:.2f}')
```

```python
# Visualización 3D de la regresión múltiple usando las dos primeras variables
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Usamos solo las dos primeras variables para graficar en 3D
x1 = X_multi[:, 0]
x2 = X_multi[:, 1]
ax.scatter(x1, x2, y_multi, color='blue', label='Datos reales')

# Crear una malla para la superficie de predicción
x1_surf, x2_surf = np.meshgrid(np.linspace(x1.min(), x1.max(), 20), np.linspace(x2.min(), x2.max(), 20))
x3_mean = np.mean(X_multi[:, 2])
X_surf = np.c_[x1_surf.ravel(), x2_surf.ravel(), np.full(x1_surf.size, x3_mean)]
y_surf = lin_reg_multi.predict(X_surf).reshape(x1_surf.shape)

ax.plot_surface(x1_surf, x2_surf, y_surf, color='red', alpha=0.5, label='Plano de regresión')

ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('y')
ax.set_title('Regresión Lineal Múltiple (Visualización 3D)')
plt.show()
```
